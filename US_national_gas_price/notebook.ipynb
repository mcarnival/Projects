```python
import pandas as pd
import seaborn as sns
import math
```

```python
df = pd.read_html("https://www.eia.gov/dnav/pet/hist/LeafHandler.ashx?n=pet&s=emm_epmr_pte_nus_dpg&f=m")[6]
```

```python
df.dropna()
ax = sns.histplot(df['Apr'])
ax.set(xlabel='Gas Price in $',
       ylabel='count',
       title='Apr Price distribution')
```

```python
import statsmodels.api as sm

df= df.dropna()
# Preparing the data
X = df['Year']
y = df['Apr']

# Adding a constant to the predictor variable set
X = sm.add_constant(X)

# Logistic Regression model
model = sm.GLM(y, X, family=sm.families.Gamma()).fit()
# Model summary
print(model.summary())
```
