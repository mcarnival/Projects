## Process
The tools I decided to use for this project was the R programming software. The reason was the processing power to modify, organize, and clean the dataset. I will also be using R to display visuals between he casual vs. membership. The data’s integrity was reviewed and attended in the prepare phase to minimize the impact of missing or unusable of the data. The steps I have taken to clean the data was to append the 12 months’ worth of data into 1 data frame or dataset to make working with other datasets easier, I dropped all the NAs that were present in the datasets. Some of the columns’ name were changed to better represent the data. Finally, I change the datatype of the time in the field, started_at, and ended_at from char to POSIXct in order to find ride length of each ride. To verify that my data is clean, I have used R importing to trim spaces, check for duplicates, convert the datatype from char to POSIXct, renamed columns to fit the data subject, and drop all blank cells. I have documented my cleaning process in the R programming software platform shown below

```r
# install packages
library(tidyverse)

# load dataset
library(readr)
bs_jan_21 <- read_csv("/kaggle/input/tripdata2021/202101-divvy-tripdata.csv")
bs_feb_21 <- read_csv("/kaggle/input/tripdata2021/202102-divvy-tripdata.csv")
bs_mar_21 <- read_csv("/kaggle/input/tripdata2021/202103-divvy-tripdata.csv")
bs_apr_21 <- read_csv("/kaggle/input/tripdata2021/202104-divvy-tripdata.csv")
bs_may_21 <- read_csv("/kaggle/input/tripdata2021/202105-divvy-tripdata.csv")
bs_jun_21 <- read_csv("/kaggle/input/tripdata2021/202106-divvy-tripdata.csv")
bs_jul_21 <- read_csv("/kaggle/input/tripdata2021/202107-divvy-tripdata.csv")
bs_aug_21 <- read_csv("/kaggle/input/tripdata2021/202108-divvy-tripdata.csv")
bs_sep_21 <- read_csv("/kaggle/input/tripdata2021/202109-divvy-tripdata.csv")
bs_oct_21 <- read_csv("/kaggle/input/tripdata2021/202110-divvy-tripdata.csv")
bs_nov_21 <- read_csv("/kaggle/input/tripdata2021/202111-divvy-tripdata.csv")
bs_dec_21 <- read_csv("/kaggle/input/tripdata2021/202112-divvy-tripdata.csv")

#combine dataset
bs_df <- bind_rows(bs_jan_21,bs_feb_21, bs_mar_21, bs_apr_21, bs_may_21, bs_jun_21, bs_jul_21, bs_aug_21,
                     bs_sep_21, bs_oct_21, bs_nov_21, bs_dec_21, id = NULL)
write.csv(bs_df,'bs_df.csv')

#clean data and converting char to time
bs_df$started_at <- as.POSIXct(bs_df$started_at)
bs_df$ended_at <- as.POSIXct(bs_df$ended_at)
bs_df <- distinct(bs_df)

#finding the difference between time and getting weekdays
ride_length<- difftime(bs_df$ended_at, bs_df$started_at, units = "mins")
weekday <- weekdays(bs_df$started_at)

#combine dataset to include the total time used from start to end 
bs_df <- cbind(bs_df, ride_length)
bs_df <- cbind(bs_df, weekday)

#drop all the NAs
bs_df <- bs_df %>% drop_na()
write.csv(bs_df,'bs_df.csv')

#changing the name of the data frame
bs_df <- rename(bs_df, rider_type = member_casual)
bs_df <- rename(bs_df, bike_type = rideable_type)
bs_df <- rename(bs_df, ride_length_mins = ride_length)

#creating a table for analyzing
bs_analyze_table <- bs_df %>% select(rider_type, bike_type, 
                                     start_station_id, end_station_id, weekday, ride_length_mins)
# Reorder the levels of the 'weekdays' variable
bs_analyze_table$weekday <- factor(bs_analyze_table$weekday, levels = c("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"))
```

## Analyze

To prepare for the analysis portion of the data, I organized the dataset by:
•	Aggregating the dataset from month to month for the year 2021 to one data frame in R
•	Converting the dates from the column ‘started_at’ and ‘ended_at’ from character type to POSIXct in order to find the difference between the time to get the amount of ride length
•	Adding in two columns for the ride length, and the weekdays column.
•	Made sure to delete any duplicate data by using the function ‘distinct().’ 
I created another data frame in the name of “bs_analysis_table “for analyzing how members and causals rider differ. The data has been narrowed down to 6 columns from the bs_df.csv to process data more smoothly. I only needed those columns to do analysis. Then I created another data frame, “bs_summary.csv” to include the descriptive statistics between the member and causal riders.

```r
#Creating bs_summary statistics

bs_summary<-bs_analyze_table %>% group_by(rider_type, weekday) %>%
  summarize(sum_time_min = sum(ride_length_mins),
            mean_ride_time =mean(ride_length_mins),
            sd_ride_time = sd(ride_length_mins))
            
# Reorder the levels of the 'weekdays' variable

bs_summary$weekday <- factor(bs_summary$weekday, levels = c("Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"))
```

After going through the numbers and making charts, I found that in chart 1, there are more riders with membership than riders without. In addition, people with membership did not ride on any docked bike. 

```r

# plot 1
plot1 <- ggplot(data = bs_analyze_table) + 
  geom_bar(mapping = aes(x =rider_type, fill = bike_type)) + 
  scale_fill_manual("bike type", values=c("#00C5CD", "#00FF00", "#FFEC8B")) +
    labs(title = "Annual Bike Ride", subtitle ="Member vs Casual",  
         caption = paste0("Data from: Jan-2021 to Dec-2021"), x="Rider Type", y="Total Rides")
plot1
```

In chart 2, causal riders had a significantly more ride length. This was very surprising because I would have not known this just by looking at the table. Also, the chart displayed in chart 2, shows more than three times increase in ride length for casual riders on Saturday and Sundays.

```r
# plot 2
plot2 <-ggplot(data = bs_summary) + geom_col(mapping = aes(x = weekday, y = sum_time_min, fill = rider_type))
plot2 <- plot2 + scale_fill_manual("Rider Type", values=c("#00C5CD", "#FFF68F")) + 
  labs(title = "Annual Weekly Ride Time", subtitle ="Member vs Casual",  caption = paste0("Data from: Jan-2021 to Dec-2021"), x="Weekday", y="Total Rides in Minutes")
plot2 <- plot2 + scale_color_discrete("Rider Type")
plot2
```

In chart 3, casual rider prefers to ride on Saturdays and Sundays. These insights tell us how causal and member differ in their riding length, causal rider use 3 options of bike type compared to just 2 options in member rider.

```r
# plot 3
plot3 <- ggplot(data = bs_analyze_table) + geom_bar(mapping = aes(x =rider_type)) + 
  facet_wrap(~weekday) +  labs(title = "Annual Weekly Total Ride", subtitle ="Member vs Casual",  
       caption = paste0("Data from: Jan-2021 to Dec-2021"), x="Weekday", y="Total Rides")
plot3
```
